```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('cost_revenue_clean.csv')
```

```python
data.describe()
```

```python
data.head()
```

### Split data

```python
X = DataFrame(data, columns=['production_budget_usd'])
y = DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

**Linear regression is going to calculate the line that minimizes the residuals. The sum of the squared residuals is the number that the linear regression will try to minimize in order to choose the best parameters for the line.**

```python
regression = LinearRegression()
regression.fit(X, y)
```

### Slope coefficient:

```python
regression.coef_   
```

**This means that there is a positive relationship between budget and revenue and that for each dollar spent on producing the movie, one should get around 3.1 dollars.**

### Intercept

```python
regression.intercept_
```

**What this intercept is telling is that a movie with a budget of zero would actually lose over 7 million dollars. This is not realistic.**

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)

# Adding the regression line here:
plt.plot(X.values, regression.predict(X), color='red', linewidth=3)

plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

**The model seems to fit the data a lot worse at the extremes.**

**Getting R_square from Regression**

Goodness of fit:

```python
regression.score(X, y)
```

**This numberis the amount of variation in film revenue that is explained by the film's budget.**
